```python
import pandas as pd
# エクセルファイルの シート名 import から読み出す
data = pd.read_excel("紀要論文___Departmental_Bulletin_Paper.xlsx",
                     sheet_name='import', index_col=None)
# ヘッダの5行目までを取り出す
data_index = data[0:4]
# 5行目までを dropする # data_body = data.drop([0,1,2,3])
# 1行目までを drop
data_body = data.drop([0])
# data_indexの1行目をカラム名に設定
columnsname = data_index.loc[0].values.tolist()
data_body.columns = columnsname
# parent,children を文字列化
data_body['parent'] = data_body['parent'].astype(str)
data_body['children'] = data_body['children'].astype(str)
# index を振りなおす(parent + childrenを4桁に揃えて)
data_body2 = data_body.assign(index=lambda x: x.parent.str.cat(x.children.str.zfill(4))).set_index('index')
data_body2
```

```python
# children が 0 の行を抽出
# A value is trying to be set on a copy of a slice from a DataFrame
data_body3 = data_body2[data_body2['children'] == '0'].copy()
# NaN のデータを　--KESSON-- で埋める
# NaNで埋めるとfloatに変換されてしまう data_body3が
data_body3.fillna('--KESSON--', inplace=True, downcast='infer')
# data_body3 で data_body2 を更新する
data_body2.update(data_body3)
data_body2
```

```python
data_body2['.metadata.path[0]']
```

```python
# tatemochi_df = data_body2.stack().reset_index()
# tatemochi_df = data_body2.melt(id_vars='children', var_name='項目', value_name='値')
# melt 残す変数（id_vars ）と melt し縦持ちにする変数（values_vars）を指定
tatemochi_df = data_body2.melt(id_vars=["children", "parent"])
tatemochi_df2 = tatemochi_df.dropna(subset=['value'])
# where()で、NaNをNoneに変換　NaNだとFloatなのでreplaceできないため。
tatemochi_df2.where(tatemochi_df2.notna(), None)
tatemochi_df2
```

```python
# 項目の[0]をchildrenの[x]で置換して、'項目2' 列を作成
# tatemochi_df2->tatemochi_df 2に戻す
# <ipython-input-5-43466514a022>:3: SettingWithCopyWarning:
# A value is trying to be set on a copy of a slice from a DataFrame.
# Try using .loc[row_indexer, col_indexer] = value instead
tatemochi_df2.loc[:,"項目2"] = tatemochi_df2.apply(
    lambda x: x.variable.replace('[0]', '[' + str(x.children) + ']', 1), axis=1)
tatemochi_df2
```

```python
print(tatemochi_df2)
```

```python
# 縦持ちを横持ちに戻す
data_body4 = tatemochi_df2.pivot(index='parent', columns='項目2', values='value')
# print(data_body4.dropna(how='all'))
data_body4

# ValueError: Index contains duplicate entries, cannot reshape
# ラベル同士の組み合わせで被りがあるとエラーが返ってきます。
# tatemochi_df2 = tatemochi_df.dropna(subset=['value'])　をしないと・・・
```

```python
# --KESSON-- の削除
# import numpy as np
# data_body5 = data_body4.replace('--KESSON--', np.nan)
# np.nan だと 数値がfloatに変換されてしまう・・・
data_body5 = data_body4.replace('--KESSON--', None)

```

```python
# metadata_path がfloat だと小数点がついて.0になっているのを Int64 に変換
# --KESSON-- の削除をして,np.nanで埋めてしまう前に、float->int->str しないといけない
# pandas.DataFrameの型を変えたいけど例外レコードがあってastypeできない時の対処法
# iloc と loc を混在させるには
# print(data_body5.iloc[3:, data_body5.columns.get_loc('.metadata.path[0]')])
metadata_path = data_body5.iloc[3:, ].filter(like='.metadata.path[').columns.tolist()
for metadata_path_column in metadata_path:
    data_body5[metadata_path_column] = pd.to_numeric(data_body5.iloc[3:, data_body5.columns.get_loc(
      metadata_path_column)], errors='coerce').astype('Int64')
```

```python
data_body5.iloc[3:, ].filter(like='.metadata.path[')
```

```python
# [DUMMY]の削除
# print(data_body5.filter(like='[DUMMY]', axis=1))
data_body6 = data_body5.drop(data_body5.filter(like='[DUMMY', axis=1),axis=1)
```

```python
# エクセルから読み込んだ段階でtimestampになっている項目を修正
# 順序とインデックス名を同時に参照するには
# data_body6.iloc[3:, data_body6.columns.get_loc('.metadata.pubdate')]
s = pd.to_datetime(
    data_body6.iloc[3:, data_body6.columns.get_loc('.metadata.pubdate')])
# print(pd.to_datetime(
#     data_body6.iloc[3:, data_body6.columns.get_loc('.metadata.pubdate')]))
s2 = s.dt.strftime('%Y-%m-%d')
print(s2)
data_body6.iloc[3:, data_body6.columns.get_loc('.metadata.pubdate')] = s2
```
